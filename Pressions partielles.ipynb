```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

#Lecture du fichiers CSV
dfPP = pd.read_csv("Pressions partielles.csv", sep=";", decimal=",")

for col in dfPP.columns[3:]:
    dfPP[col] = pd.to_numeric(dfPP[col], errors="coerce")
    
plt.figure(figsize=(8, 5))

for col in dfPP.columns[3:]:
    plt.plot(dfPP["t_h"], dfPP[col], label=col)

plt.xlabel("Time (h)")
plt.ylabel("Partial pressure (mbar)")
plt.yscale("log")
plt.legend(loc = "upper left")
plt.grid()
plt.tight_layout()
plt.show()
```
